# Bold and Italic Text:

You can use `<b>` tags, '**' i.e. 'double asterisk' or '__' i.e. 'double underscore' to get bold text with the following syntax.

<b>This is bold text </b> and this is NOT bold

You can use `<i>` tags, '*' i.e., single asterisk or '_' i.e., single underscore to get the italic text for the following syntax.

<i>This is <b>italic AND bold</b> text </i>

# Horizontal Line:

You can obtain a horizontal line by using Markdown '---' three hyphens or Markup tags `<hr>`

<hr>

# Ordered List:

The Ordered List is the numbered list which is obtained by starting `<ol>` tag and ending with `</ol>` tag with the required item in between `<li>` and `</li>` tags. <br>The tag, i.e., 'ol,' is the short form for an ordered list and 'li' is the short form for the list item. <br>For example, you can see the Ordered List below containing the item for the grocery list.

<ol>
<li>Fish</li>
<li>Eggs</li>
<li>Cheese</li>
</ol>

---

# Unordered List:

The Unordered list is a bullet list which is obtained by using the `<ul>` tag and ending with the `</ul>` tag, see the example below:

<ul>
<li>Fish</li>
<li>Eggs</li>
<li>Cheese</li>
</ul>
